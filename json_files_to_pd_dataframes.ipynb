### Step 1: Import packages

```python
import pandas as pd
import json
import os
```

### Step 2: Load the JSON file containing the scraped data from the Polish website

```python
with open(file=os.path.join(os.getcwd(), "product_page.json"), mode="r") as f:
    df = json.load(f)
    df = pd.DataFrame(df)
    f.close()
```
